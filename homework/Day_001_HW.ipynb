```python
# 載入基礎套件與代稱
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mean_absolute_error(y, yp):
    """
    計算 MAE
    Args:
        - y: 實際值
        - yp: 預測值
    Return:
        - mae: MAE
    """
    mae = MAE = sum(abs(y - yp)) / len(y)
    return mae

# 定義 mean_squared_error 這個函數, 計算並傳回 MSE
def mean_squared_error(y, y_hat):
    mse = sum(pow(y-y_hat, 2)) / len(y)
    return mse
```

```python
w = 2
b = 1
x_lin = np.linspace(0, 50, 51)

# 用linear regression來預測，看成預測值
y = w*x_lin + b

# 建立資料，以linear regression作為基礎再加入一些亂數，讓資料浮動
y1 = w*(x_lin + np.random.rand(51)*5) + b    # np.random.rand() 這種隨機取樣是從0~1隨機取樣
y2 = w*(x_lin + np.random.randn(51)*5) + b  # np.random.randn() 這種隨機取樣是常態分配隨機取樣

fig = plt.figure()
ax = plt.subplot(3,1,1)
ax1 = plt.subplot(3,1,2)
ax2 = plt.subplot(3,1,3)
ax.scatter(x_lin, y, c='b', alpha=0.8)
ax1.scatter(x_lin, y1, c='r', alpha=0.8, edgecolors='none')
ax2.scatter(x_lin, y2, c='g', alpha=0.8, edgecolors='none')

ax.legend(["no random"])
ax1.legend(["random"])
ax2.legend(["randn"])

plt.show()
```

```python
# 與範例相同, 不另外解說
w = 3
b = 0.5
x_lin = np.linspace(0, 100, 101)
y = (x_lin + np.random.randn(101) * 5) * w + b

plt.plot(x_lin, y, 'b.', label = 'data points')
plt.title("Assume we have data points")
plt.legend(loc = 2)
plt.show()
```

```python
# 與範例相同, 不另外解說
y_hat = x_lin * w + b
plt.plot(x_lin, y, 'b.', label = 'data')
plt.plot(x_lin, y_hat, 'r-', label = 'prediction')
plt.title("Assume we have data points (And the prediction)")
plt.legend(loc = 2)
plt.show()
```

```python
# 執行 Function, 確認有沒有正常執行
MSE = mean_squared_error(y, y_hat)
MAE = mean_absolute_error(y, y_hat)
print("The Mean squared error is %.3f" % (MSE))
print("The Mean absolute error is %.3f" % (MAE))
```
