### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets

digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
print(X.shape)

# 可以看到 y確實只有分三類
print(set(y))
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
import numpy as np
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
pca = PCA(n_components=3)
X = pca.fit_transform(X)
```

```python
fig = plt.figure(num=1, figsize=(6, 8))
ax = Axes3D(fig, rect=[0, 0, 0.9, 0.7], elev=48, azim=134)
ax.dist = 11
ax.scatter(X[:, 2], X[:, 0], X[:, 1], c=y, edgecolor='k')

label_title = [[0, 'class0'], [1, 'class1'], [2, 'class2']]
for i, name in label_title:
    ax.text3D(X[y==i, 2].mean(),
              X[y==i, 0].mean(),
              X[y==i, 1].mean(), name, 
              bbox=dict(alpha=.7, edgecolor='w', facecolor='w'))
    
```
