```python
# 載入套件
import time
import warnings

import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler
from itertools import cycle, islice
```

```python
# 設定 2D 樣板資料
n_samples = 1500
random_state = 100

# 生成 雙半月 資料點
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)

# 生成 緊密三群 資料點
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)

# 生成 2維均勻分布 資料點
no_structure = np.random.rand(n_samples, 2), None
```

# 作業
* 試著完成剩下三種 2D 樣板資料的繪圖, 並觀察有何不同

```python
plt.figure(figsize=(20, 16))
plot_num = 1
default_base = {'n_neighbors': 10, 'n_clusters': 3}
datasets = [(noisy_moons, {'n_clusters': 2}), (blobs, {}), (no_structure, {})]

# 執行各種樣板資料的繪圖迴圈
for i, (dataset, algo_params) in enumerate(datasets):
    params = default_base.copy()
    params.update(algo_params)
    X, y = dataset
    X = StandardScaler().fit_transform(X)
    #print(params)
    #print(X)
    
    ward = cluster.AgglomerativeClustering(n_clusters=params['n_clusters'], 
                                           linkage='ward')
    comp = cluster.AgglomerativeClustering(n_clusters=params['n_clusters'], 
                                           linkage='complete')
    avg = cluster.AgglomerativeClustering(n_clusters=params['n_clusters'], 
                                          linkage='average')
    estimators = [('Average', avg),
                  ('complete', comp),
                  ('ward', ward)]
    
    for name, est in estimators:
        t0 = time.time()
        est.fit(X)
        t1 = time.time()
        if hasattr(est, 'labels_'):
            y_pred = est.labels_.astype(np.int)
        else:
            y_pred = est.predict(X)
        
        plt.subplot(len(datasets), len(estimators), plot_num)
        
        if i == 0:
            plt.title(name, size=18)

        colors = np.array(list(islice(cycle(['#377eb8', '#ff7f00', '#4daf4a',
                                             '#f781bf', '#a65628', '#984ea3',
                                             '#999999', '#e41a1c', '#dede00']),
                                      int(max(y_pred) + 1))))
        
        plt.scatter(X[:, 0], X[:, 1], s=10, color=colors[y_pred])
        plt.xlim(-2.5, 2.5)
        plt.ylim(-2.5, 2.5)
        plt.xticks(())
        plt.yticks(())
        plt.text(.99, .01, ('%.2fs' % (t1 - t0)).lstrip('0'), transform=plt.gca().transAxes, size=15,
                 horizontalalignment='right')
        plot_num += 1
plt.show()
```
